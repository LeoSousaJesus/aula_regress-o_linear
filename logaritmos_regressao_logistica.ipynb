# DefiniÃ§Ã£o:

### Um logaritmo Ã© a operaÃ§Ã£o inversa da exponenciaÃ§Ã£o. Enquanto a exponenciaÃ§Ã£o eleva um nÃºmero a uma potÃªncia, o logaritmo encontra a potÃªncia necessÃ¡ria para obter um determinado valor.<br><br>

#### Tipos de Logaritmos:

1. Logaritmo Natural (ln): Baseado no nÃºmero e (aproximadamente 2,718).<br>

2. Logaritmo Comum (log): Baseado no nÃºmero 10.<br>

3. Logaritmo BinÃ¡rio (log2): Baseado no nÃºmero 2.<br>
<br>

#### Propriedades:

1. InversÃ£o: log(a^b) = b * log(a)
2. Produto: log(a * b) = log(a) + log(b)
3. Quociente: log(a / b) = log(a) - log(b)
4. PotÃªncia: log(a^b) = b * log(a)
<br><br>

#### AplicaÃ§Ãµes:

- CÃ¡lculo de taxas de crescimento e decrescimento.
- ResoluÃ§Ã£o de equaÃ§Ãµes exponenciais.
- AnÃ¡lise de dados em escalas logarÃ­tmicas.
- MatemÃ¡tica, fÃ­sica, quÃ­mica, geografia, computaÃ§Ã£o, engenharia, economia e estatÃ­stica.
<br><br>

#### Exemplos:

1. log(100) = 2, pois 10 elevado a 2 = 100.
2. ln(e) = 1, pois e^1 = e.
3. log2(8) = 3, pois 2^3 = 8.
<br><br>

####Conceitos relacionados:

- ExponenciaÃ§Ã£o.
- FunÃ§Ãµes exponenciais.

# A funÃ§Ã£o logarÃ­tmica Ã© fundamental na mediÃ§Ã£o da magnitude de terremotos, pois permite expressar valores extremamente grandes em uma escala mais tratÃ¡vel. Aqui estÃ¡ como ela Ã© utilizada:

## **Escala Richter**

### A escala Richter serve para medir a magnitude de terremotos. Ela utiliza a funÃ§Ã£o logarÃ­tmica para calcular a magnitude **(M)** com base na amplitude das ondas sÃ­smicas registradas por sismÃ³grafos.
<br>

#**M = log10(A) + B**<br><br>


### Onde:

- M Ã© a magnitude do terremoto
- A Ã© a amplitude das ondas sÃ­smicas
- B Ã© um fator constante que depende da localizaÃ§Ã£o e do tipo de terremoto
<br><br>

#### Escala de Magnitude de Momento

A escala de magnitude de momento (Mw) Ã© outra medida utilizada para terremotos grandes. Ela tambÃ©m emprega a funÃ§Ã£o logarÃ­tmica:<br><br>

#**Mw = 2/3 log10(M0) - 10.7**<br><br>

###Onde:

- Mw Ã© a magnitude de momento
- M0 Ã© o momento sÃ­smico (produto da Ã¡rea de ruptura, deslizamento mÃ©dio e mÃ³dulo de elasticidade)<br><br>

#### CaracterÃ­sticas da funÃ§Ã£o logarÃ­tmica na mediÃ§Ã£o de magnitude:<br><br>

1. Reduz valores extremos: A funÃ§Ã£o logarÃ­tmica compressa valores grandes em uma escala menor.
2. Permite comparaÃ§Ã£o: A escala logarÃ­tmica permite comparar magnitudes de terremotos de diferentes tamanhos.
3. Linearidade: A relaÃ§Ã£o entre a magnitude e a amplitude das ondas sÃ­smicas Ã© aproximadamente linear na escala logarÃ­tmica.<br><br>

####**Exemplos:**

- Terremoto de 1906 em SÃ£o Francisco (M 7,9): log10(A) â 7,9
- Terremoto de 2011 no JapÃ£o (M 9,1): log10(A) â 9,1

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

#RegressÃ£o LogÃ­stica com o dataset de vendas
# Carregando os dados de vendas
dados = pd.read_csv('https://raw.githubusercontent.com/romulosilvestre/rl_01/refs/heads/main/vendas.csv')
X_vendas = dados['Idade'].values.reshape(-1, 1)  # Transformando em uma matriz 2D
y_vendas = dados['Comprou']

# Criando e treinando o modelo
modelo_vendas = LogisticRegression()
modelo_vendas.fit(X_vendas, y_vendas)

# Gerando probabilidades para o grÃ¡fico
X_plot = np.linspace(X_vendas.min(), X_vendas.max(), 100).reshape(-1, 1)
y_prob = modelo_vendas.predict_proba(X_plot)[:, 1]

# GrÃ¡fico de dispersÃ£o para os dados de vendas
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)  # Adicionar um subplot para a primeira visualizaÃ§Ã£o
plt.scatter(X_vendas, y_vendas, s=2)
plt.plot(X_plot, y_prob, color='red', linewidth=2)
plt.title("Compras por idade")
plt.xlabel("Idade")
plt.ylabel("Probabilidade de compra")
plt.grid(True)

#RegressÃ£o LogÃ­stica com o dataset Iris
# Carregando os dados Iris
iris = load_iris()
X_iris = iris.data[:, :2]  # Utilizando apenas duas caracterÃ­sticas (comprimento e largura da sÃ©pala)
y_iris = iris.target

# Dividindo os dados em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X_iris, y_iris, test_size=0.2, random_state=42)

# Criando e treinando o modelo
modelo_iris = LogisticRegression(max_iter=200)
modelo_iris.fit(X_train, y_train)

# Fazendo previsÃµes
y_pred = modelo_iris.predict(X_test)

# Avaliando o modelo
accuracy = accuracy_score(y_test, y_pred)
print(f"AcurÃ¡cia do modelo Iris: {accuracy:.2f}")

# GrÃ¡fico de dispersÃ£o para o dataset Iris
plt.subplot(1, 2, 2)  # Segundo subplot para visualizaÃ§Ã£o do Iris
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap='viridis', marker='o', edgecolor='k', label='Treino')
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_pred, cmap='plasma', marker='x', edgecolor='k', label='Teste')
plt.title(f"Iris dataset (AcurÃ¡cia: {accuracy:.2f})")
plt.xlabel("Comprimento da sÃ©pala")
plt.ylabel("Largura da sÃ©pala")
plt.legend()
plt.grid(True)

# Exibindo os grÃ¡ficos
plt.tight_layout()
plt.show()
```
