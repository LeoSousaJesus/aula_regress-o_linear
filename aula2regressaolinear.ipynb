# Treino e Teste

```python
# preparar os dados
X = df.iloc[:,0].values
```

```python
print(f"Anos de Experiência:{X}")
```

```python
y = df.iloc[:,1].values
```

```python
print(f"Salário:{y}")
```

# Separando os dados (Treino e Teste)

image.png

```python
from sklearn.model_selection import train_test_split
```

## train_test_split: separa treino e teste (validação cruzada)

```python
# Dados -X
# Dados de Treino - X_train
# Dados de Teste - X_test
#------------------------------------------------------------------------------------------------------------------------------------
# Classes (rótulos) - y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=1/3, random_state=0)
```

```python
# Importar o algoritmo
from sklearn.linear_model import LinearRegression
```

```python
# regressor é um objeto da classe LinearRegression()
# o objeto vai chamar métodos ( fit() predict() )

regressor = LinearRegression()
```

```python
# treinando
X_train = X_train.reshape(-1, 1)

y_train = y_train.reshape(-1, 1)
```

```python
# Sorria vc está sendo treinado
regressor.fit(X_train, y_train)
```

```python
X_test = X_test.reshape(-1, 1)
```

```python
y_pred = regressor.predict(X_test)
```

```python
y_pred
```

```python
# Finalizando com um Gráfico
import matplotlib.pyplot as plt
```

```python
# Visualizar os resultados de treino
plt.figure(figsize=(15,8))
plt.plot(X, y, 'Dr')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Salário x Experiência ')
plt.xlabel('Anos de Experiência')
plt.ylabel('Salário')
plt.show()
```
