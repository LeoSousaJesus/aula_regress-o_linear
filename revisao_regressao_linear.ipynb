Regressão linear é um método usado para encontrar a relação entre duas variáveis, onde uma é a variável independente (entrada) e a outra é a variável dependente (saída). Basicamente, ele busca uma linha reta que melhor aproxima os pontos de dados em um gráfico, ajudando a prever o valor da saída com base no valor da entrada

# Qual a diferença entre regressão linear e regressão lógistica?



1.   Enquanto a linear prever um valor, a lógistica uma probabilidade.
2.   A probabilidade esta entre 0 e 1 (variável categórica)
3.   A aplicação da lógistica frequentemente é binária
4.   Utiliza-se variáveis númericas contínuas na regressão linear, na lógistica também, em relação as variáveis independentes.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Suponha que você tenha um DataFrame chamado 'dataset' com colunas 'Experiencia' e 'Salario'
# dataset = pd.read_csv('seu_arquivo.csv') # Substitua 'seu_arquivo.csv' pelo nome do seu arquivo

# Se você já tiver os dados em variáveis X e y, pode pular essa etapa.
# X = dataset['Experiencia'].values.reshape(-1, 1)
# y = dataset['Salario'].values

# Dividir os dados em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

# Criar o modelo de Regressão Linear
model = LinearRegression()

# Treinar o modelo
model.fit(X_train, y_train)

# Fazer previsões com os dados de teste
y_pred = model.predict(X_test)

# Plotar o gráfico
plt.scatter(X, y, color='red')
plt.plot(X_train, model.predict(X_train), color='blue')
plt.title('Salário vs Experiência')
plt.xlabel('Experiência')
plt.ylabel('Salário')
plt.show()

# Prever o salário para um novo valor de experiência
novo_X = [[10]]  # Prever o salário para 10 anos de experiência
y_pred_novo = model.predict(novo_X)
print("Salário previsto para 10 anos de experiência:", y_pred_novo[0])
```

```python
# passo 1 - importando pandas, matplotlib e numpy
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# importando o módulo linear_model da biblioteca scikit-learn
# importando a classe LogisticRegression
from sklearn.linear_model import LogisticRegression
```

```python
#passo 2 - carregando os dados
dados = pd.read_csv('https://raw.githubusercontent.com/romulosilvestre/rl_01/refs/heads/main/vendas.csv')
```

```python
# passo 3 - definindo X e y
X = dados['Idade']
y = dados['Comprou']
```

```python
model = LogisticRegression()
```

```python
model.fit(X, y)
```

```python
# passo 4 - definindo gráfico de dispersão com espessura s=2
X = np.array(X).reshape(-1, 1)  # Converter X em um array 2D

# Gerar pontos para a linha de regressão
X_plot = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
y_prob = model.predict_proba(X_plot)[:, 1]  # Probabilidade da classe 1

# Gráfico de dispersão
plt.scatter(X, y, s=2)
plt.title("Compras por idade")
plt.xlabel("Idade")
plt.ylabel("Comprou")

# Adicionar a linha de regressão logística
plt.plot(X_plot, y_prob, color='red', linewidth=2)

plt.show()
```
